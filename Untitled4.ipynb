```python
from scipy.sparse import csr_matrix
```

```python
video_sparse = csr_matrix(PivotTable)
```

```python
video_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm ='brute')
```

```python
model.fit(video_sparse)
```

```python
distance,suggestion = model.kneighbors(PivotTable.iloc[23,:].values.reshape(1,-1), n_neighbors = 5)
```
